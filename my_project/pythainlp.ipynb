```python
import numpy as np
import pandas as pd
from apyori import apriori
from mlxtend.frequent_patterns import apriori, association_rules 
```

```python
df = pd.read_csv('/Users/Ratchasit/Downloads/test_user_post_3.csv',error_bad_lines=False)
```

```python
df.head()
```

```python
df_cut = df[['unique_id','post_id']]
df_cut['quantity'] = 1
```

```python
basket_France = (df_cut
          .groupby(['unique_id', 'post_id'])['quantity'] 
          .sum().unstack().reset_index().fillna(0) 
          .set_index('unique_id'))
```

```python
def hot_encode(x): 
    if(x<= 0): 
        return 0
    if(x>= 1): 
        return 1
```

```python
basket_encoded = basket_France.applymap(hot_encode) 
basket_France = basket_encoded
```

```python
# Building the model 
frq_items = apriori(basket_France, min_support = 0.03, use_colnames = True) 
  
# Collecting the inferred rules in a dataframe 
rules = association_rules(frq_items, metric ="lift", min_threshold = 1) 
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False]) 
rules.head()
```
